```python
# let's walk through those files
import os
def walk_through_files(path, file_extension='.html'):
   list_of_files = []
   for (dirpath, dirnames, filenames) in os.walk(path):
      for filename in filenames:
         if filename.endswith(file_extension): 
            list_of_files.append(os.path.join(dirpath, filename))
   return list_of_files
```

```python
list =  walk_through_files("Workload Estimation/Raw Datasets/fastStorage/2013-8/", ".csv")
```
